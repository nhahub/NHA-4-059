```python
import os
import json
import random

base_path = r"E:\DIPE\Project_DEPI\data\raw"

classes = [
    "garbage_truck", "moving_van", "pickup", "trailer_truck",
    "beer_truck", "fire_engine", "tow_truck", "minivan"
]

random.seed(42) 

splits = {}

for class_name in classes:
    folder = os.path.join(base_path, class_name)
    images = sorted(os.listdir(folder))
    random.shuffle(images)
    
    total = len(images)
    train_end = int(total * 0.70)
    val_end = int(total * 0.85)
    
    splits[class_name] = {
        "train": images[:train_end],
        "val": images[train_end:val_end],
        "test": images[val_end:]
    }
    
    print(f"✓ {class_name}: train={len(splits[class_name]['train'])} | val={len(splits[class_name]['val'])} | test={len(splits[class_name]['test'])}")


splits_path = r"E:\DIPE\Project_DEPI\data\splits.json"
with open(splits_path, 'w') as f:
    json.dump(splits, f, indent=2)

print("\n✓ splits.json اتحفظ!")
```

```python
import os
import json
import random

base_path = r"E:\DIPE\Project_DEPI\data\raw"


target_classes = ["beer_truck", "moving_van"]

logo_candidates = []

for class_name in target_classes:
    folder = os.path.join(base_path, class_name)
    images = sorted(os.listdir(folder))
    

    for img in images:
        logo_candidates.append({
            "filename": img,
            "class": class_name,
            "path": os.path.join(folder, img)
        })

random.seed(42)
selected = random.sample(logo_candidates, 100)


save_path = r"E:\DIPE\Project_DEPI\data\logo_candidates.json"
with open(save_path, 'w') as f:
    json.dump(selected, f, indent=2)

print(f"✓ تم تحديد {len(selected)} صورة كـ logo candidates")
print(f"✓ logo_candidates.json اتحفظ!")

```

```python
import os
from datasets import load_dataset
from PIL import Image


token = "YOUR_HF_TOKEN_HERE"

base_path = r"E:\DIPE\Project_DEPI\data\raw"

classes = {
    "garbage_truck": 573,
    "moving_van": 675,
    "pickup": 717,
    "trailer_truck": 867,
    "beer_truck": 258,
    "fire_engine": 555,
    "tow_truck": 864,
    "minivan": 656
}

# عمل الفولدرات
for class_name in classes:
    os.makedirs(os.path.join(base_path, class_name), exist_ok=True)

print("جاري تحميل ImageNet...")
dataset = load_dataset(
    "ILSVRC/imagenet-1k",
    split="train",
    streaming=True,
    token=token
)

counters = {name: 0 for name in classes}
target = 200

for item in dataset:
    label = item['label']
    for class_name, class_label in classes.items():
        if label == class_label and counters[class_name] < target:
            img = item['image']
            if img.mode != 'RGB':
                img = img.convert('RGB')
            save_path = os.path.join(base_path, class_name, f"{class_name}_{counters[class_name]:04d}.jpg")
            img.save(save_path)
            counters[class_name] += 1
            print(f"✓ {class_name}: {counters[class_name]}/{target}")
            break

    if all(v >= target for v in counters.values()):
        print("\nاكتمل التحميل!")
        break
```
